Outlier Removal

*horizontal/vertical distance to hydrology, horizontal distance to roadways, horizontal distance to firepoints; distance to hydrology, total hillshade*

```python
# identify outliers with interquartile range
import numpy as np
from numpy import percentile
# calculate interquartile range
q25 = []
q75 = []
features = ['Horizontal_Distance_To_Hydrology', 'Vertical_Distance_To_Hydrology', 'Horizontal_Distance_To_Roadways', 'Horizontal_Distance_To_Fire_Points', 
           'Distance_To_Hydrology', 'Total_Hillshade']
for i in features:
    q25.append(percentile(fe_train_data[i], 25))
    q75.append(percentile(fe_train_data[i], 75))
q25 = np.array(q25)
q75 = np.array(q75)
cut_off = (q75 - q25) * 1.5
lower=  q25 - cut_off
upper = q75 + cut_off
```

```python
# identify outliers
outliers = []
for i in range(len(features)):
    outliers.append(np.argwhere((fe_train_data[features[i]].values<lower[i]) |  (fe_train_data[features[i]].values>upper[i])).flatten( 'K'))
```

```python
outliers = np.concatenate(outliers).ravel()
```

```python
from collections import Counter
dic = Counter(outliers)
d={}
for key, value in dic.items():
    d[key] = value
def getKey(dct,value):
     return [key for key in dct if (dct[key] >= value)]
outlier_index = getKey(d, 3)
```

```python
fe_train_data_outlier = fe_train_data.drop(fe_train_data.index[outlier_index])
```
